## Distance btw all the bus stands

```python
path = "D:\pune\stops.txt"
stops = pd.read_csv(path)
```

```python
stops.head(2)
```

```python
id_map = defaultdict(lambda: 0)
cord = []


for i in range(len(stops)):
    id_map[i] = stops.loc[i]["stop_id"]
    cord.append([stops.loc[i]["stop_lat"],
                stops.loc[i]["stop_lon"]])
```

```python
real_fake = {}
fake_real = {}
for i in range(len(stops)):
    real_fake[stops.loc[i]["stop_id"]] = i
    fake_real[i] = stops.loc[i]["stop_id"]
```

```python
l = [[0 for i in range(len(stops))] for j in range(len(stops))]
for i in range(len(stops)):
    for j in range(len(stops)):
        l1 = tuple(cord[i])
        l2 = tuple(cord[j])
        l[i][j] = float(hs.haversine(l1, l2))
```

```python
file = open('DistanceMatrix.txt', 'w')
for i in l:
    file.write(','.join([str(j) for j in i])+"\n")
file.close()
```
